## Image of the "saddle" function

The two-dimensional "saddle" function is a function given by the product of its coordinates `r` and `c`.
Note that, by implementing it this way, it is possible to make the Python/NumPy code look
very close to the mathematical equation, placed below.

Let's generate a saddle function, where the values for the rows will be integer values between
-75 and 75 and the values for the columns, integers in the range [-100,100]:

$$ f(r,c) = r \ c $$
$$ \text{to} \r \in [-75.75] $$
$$ c \in [-100,100]$$
   
   
In the following example, the `arange` function is used to generate the coordinate vectors. 

```python
r,c = np.meshgrid(np.arange(-75,75), np.arange(-100,100), indexing='ij')
f = r * c
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.title('Saddle point')
plt.imshow(f, cmap='gray')
```
