# Numpy: Index and meshgrid functions

The ``indices`` and ``meshgrid`` functions are extremely useful in generating synthetic images and their learning also allows
understand the advantages of matrix programming, avoiding the sequential scanning of the image that is very common in programming in the C language.

## Index operator in small numeric examples

The ``indices`` function receives as parameters a tuple with the dimensions (H,W) of the matrices to be created. In the following example, we are
generating matrices of 5 rows and 10 columns. This function returns a tuple of two matrices that can be obtained by doing their assignments
as in the following example where we create the matrices ``r`` and ``c``, both of size (5,10), that is, 5 rows and 10 columns:

```python
import numpy as np
r,c = np.indices( (5, 10) )
print('r=\n', r)
print('c=\n', c)
```

Note that the matrix ``r`` is a matrix where each element is its row coordinate and the matrix ``c`` is a matrix where each element is
its column coordinate. This way, any matrix operation done with ``r`` and ``c``, in reality you are processing the
matrix coordinates. Thus, it is possible to generate several synthetic images from a function of their coordinates.

As NumPy processes matrices directly, without the need to do an explicit ``for``, the program's notation is very simple
and efficiency too. The only drawback is the use of memory to calculate the index matrices ``r`` and ``c``. We will
See later that this can be minimized.

For example, let the function be the sum of its coordinates $f(r,c) = r + c$:

```python
f = r + c
print('f=\n', f)
```

Or even the difference function between the row and column coordinates $f(r,c) = r - c$:

```python
f = r - c
print('f=\n', f)
```

Or even the function $f(r,c) = (r + c) \% 2$ where % is the modulus operator. This function returns 1 if the sum of the coordinates is odd and 0 otherwise.
It's a chessboard-style image of values 0 and 1:

```python
f = (r + c) % 2
print('f=\n', f)
```

Or even the function of a straight line $ f(r,c) = (r = \frac{1}{2}c)$:

```python
f = (r == c//2)
print('f=\n', f)
print('f=\n',f.astype(np.int16))
```

Or even the parabolic function given by the sum of the square of its coordinates

$$ f(r,c) = r^2 + c^2 $$:

```python
f = r**2 + c**2
print('f=\n', f)
```

   
Or even the function of a circle with radius 4, with center at (0,0) $f(r,c) = (r^2 + c^2 < 4^2)$:

```python
f = ((r**2 + c**2) < 4**2)
print('f=\n', f * 1)
```

## Meshgrid
   
The ``meshgrid`` function is similar to the ``indices`` function as
previously, however, while ``indices`` generates non-negative integer coordinates from a ``shape(H,W)``,
``meshgrid`` generates matrix values from two vectors of any real values, one for the rows and one for the columns.

See below a small numerical example. For the ``meshgrid`` to be compatible with our (rows, columns) convention, you must
use the ``indexing='ij'`` parameter.

```python
import numpy as np
r, c = np.meshgrid( np.array([-1.5, -1.0, -0.5, 0.0, 0.5]), 
                    np.array([ -20,  -10,    0,  10,  20, 30]), indexing='ij')
print('r=\n',r)
print('c=\n',c)
```

## Generating vectors with linspace

The ``linspace`` function generates a floating point vector by receiving the parameters initial value, final value and number of points in the vector.
This way it is often used to generate parameters for the ``meshgrid``.
 
Repeating the same values as the previous example, but using ``linspace``. Note that the first vector has 5 points,
starting with value -1.5 and the final value is 0.5 (inclusive). The second vector has 6 points, starting from -20 to 30:

```python
rows = np.linspace(-1.5, 0.5, 5)
cols = np.linspace(-20, 30, 6)

print('rows:', rows)
print('cols:', cols)
```

Using the two vectors generated by ``linspace`` in ``meshgrid``:   

```python
r, c = np.meshgrid(rows, cols, indexing='ij')
print('r = \n', r)
print('c = \n', c)
```

We can now generate a matrix or image that is a function of these values. For example being their product:

```python
f = r * c
print('f=\n', f)
```

## Example of sync image generation with meshgrid

In this example, we generate the image of the $ sinc(r,c)$ function in two dimensions, in the vertical ranges from -5 to 5 and in
horizontal from -6 to 6. The sinc function is a trigonometric function that can be used for filtering.

The equation is given by:

$$ sinc(r,c) = \frac{\sin(r^2 + c^2)}{r^2 + c^2}, \text{para\ } -5 \leq r \leq 5, - 6 \leq c \leq 6 $$

At the origin, both r and c are zeros, resulting in division by zero. However, according to limit theory, $\frac{sin(x)}{x}$ is
equals 1 when $x$ is equal to zero.
One way to obtain this in floating point is to add an *epsilon* to both the numerator and the denominator, which is the
smallest floating point value. Epsilon can be obtained by the ``np.spacing`` function.

```python
e = np.spacing(1) # epsilon to avoid 0/0
rows = np.linspace(-5.0, 5.0, 150) # row coordinates
cols = np.linspace(-6.0, 6.0, 180) # column coordinates
r, c = np.meshgrid(rows, cols, indexing='ij') # Numpy style coordinate grid
z = np.sin(r**2 + c**2 + e) / (r**2 + c**2 + e) # epsilon is added to avoid 0/0
plt.imshow(normalize(z))
```

```python
plt.imshow(z)
plt.colorbar()
```

## Example of sync image generation with indices

Another way to generate the same image, using the ``indices`` function, is to process the
indices in order to generate the same values relative to the regularly spaced grid
above, as illustrated below:

```python
n_rows = len(rows)
n_cols = len(cols)
r,c = np.indices((n_rows,n_cols))
r = -5. + 10.*r.astype(float)/(n_rows-1)
c = -6. + 12.*c.astype(float)/(n_cols-1)
zi = np.sin(r**2 + c**2 + e) / (r**2 + c**2 + e) # epsilon is addes to avoid 0/0
plt.imshow(normalize(zi))
```

Checking that the two functions are equal:

```python
print('Maximum difference between z and zi?', abs(z - zi).max())
```

## For advanced users

In reality, the `indices` function returns a single n-dimensional array with one more dimension than indicated by the
shape used as a parameter. Thus, when ``r,c = np.indices((rows,cols))`` is done, r is assigned to element 0 and
c is assigned to element 1 of the ndarray. In the case of ``meshgrid``, it returns as many arrays as the number
of vectors passed as a parameter to ``meshgrid``.
