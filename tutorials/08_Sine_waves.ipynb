# 1D and 2D sine waves

## 1D Sine Wave

For a better interpretation of the Fourier Transform, it is important to intuitively understand the sinusoidal function. Many oscillatory processes can be described by a periodic function in the form of a sinusoid: $$ f(x) = sin(\frac {2\pi}T.x + \varphi)$$

This is a continuous function that crosses the $x$ axis at the points $(k\pi,0)$ . If we now think about the discrete domain, the generic sinusoidal function is given by: $$ f(x) = A.sin(\frac {2\pi}N.u.x + \varphi)$$

where:

- $A$ is the amplitude of the signal.
- $N$ is the number of samples.
- $x$ ranges from [0 to N-1].
- $u$ is the number of cycles in the sample.

and still:

- $T= \frac Nu$ is the number of pixels in the period.
- $f= \frac uN$ is the number of cycles per pixel.

The function we will use to generate the sine waves is cos. It is capable of generating 1D and 2D cosine waves. We must remember that if we want to generate a sinusoid from the function that generates cosines, just do $$cos(x)=sin(x+\frac\pi2)$$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import sys
sys.path.append( '../master' )
from function import cos

A = 5
N = (100,1)
T = 50
T2 = 25
theta = np.pi/2
phi = np.pi/2
f = A*cos.cos(N, T, theta, 0)
f1 = A*cos.cos(N, T, theta, phi)
f2 = A*cos.cos(N, T2, theta, 0)
f3 = A*cos.cos(N, T2, theta, phi)
x = np.linspace(-np.pi,np.pi,100)


plt.figure(1, figsize=(12,2))

plt.subplot(1,4,1)
plt.title('T=50, theta = 0')
plt.plot(x,f)

plt.subplot(1,4,2)
plt.title('T=50, theta = 90')
plt.plot(x,f1)

plt.subplot(1,4,3)
plt.title('T = 25, theta = 0')
plt.plot(x,f2)

plt.subplot(1,4,4)
plt.title('T = 25, theta = 90')
plt.plot(x,f3)
```

## 2D Sine Wave

Now imagine that instead of a 1D sinusoid we have a 2D sinusoid, forming an image. The equations in the 2 D case are: $$ f(r,c) = cos(2\pi(\frac 1T_r.r + \frac 1T_c.c) + \varphi)$$ or even $$ f(r,c ) = cos(2\pi(\frac vH.r + \frac uW.c) + \varphi)$$ where $$T_r = \frac T{sin(\theta)}$$ and $$T_c = \frac T{cos(\theta)}$$ Furthermore $$u = \frac W{T_c}$$ and $$v = \frac H{T_r}$$ where:

- $\theta$ is the direction of the sine wave
- $T$ is the period of the wave, in number of pixels
- $T_r$ and $T_c$ are the periods or wavelengths in the vertical and horizontal directions, respectively, in number of pixels.
- $H$ and $W$ are the number of rows and columns in the image, respectively.
- $v$ and $u$ are the normalized frequencies in the horizontal and vertical directions, respectively, in cycles per image size.

## Visualizing a 2d cosine

Let us then consider a discrete 2D cosine wave of an image with 256 columns and 128 rows, with a wavelength of 100 pixels at 45 degrees. In this case, $T=100$, $H=128$ , $W=256$, $\theta = 45$ degrees.

```python
H = 128
W = 256
s = (H,W)
T = 100
theta = 45 * np.pi/180
f = cos.cos(s, T, theta, 0)
plt.imshow(f, cmap='gray')
```

```python
T_r = T/np.sin(theta)
T_c = T/np.cos(theta)
v = H/T_r
u = W/T_c
print('T_r:', T_r)
print('T_c:', T_c)
print('v:', v)
print('u:', u)
```
