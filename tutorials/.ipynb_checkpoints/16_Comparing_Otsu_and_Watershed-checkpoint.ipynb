# Comparing Otsu and by Watershed

Let's continue exploring segmentation techniques and understand that to achieve good segmentation, it is often necessary to pre-process the images (calculate the gradient, for example). Furthermore, you need to know how to choose the best value for each method parameter, which may not be trivial. Some techniques are easier to use, but may have lower performance. Others may be superior, but with more adjustment parameters.

### Segmentation using the Otsu Thresholding method

For some segmentation problems, simple thresholding techniques may be sufficient. In the following example, we will try to solve the segmentation of *cookies* using the Otsu thresholding method, both global and local.

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline

from skimage import measure
from skimage.filters.rank import otsu
from skimage.filters import threshold_otsu, sobel

import skimage.morphology as mm
from skimage.segmentation import watershed, join_segmentations
from skimage.segmentation import mark_boundaries
```

```python
f_pil = Image.open('../figures/cookies.pgm')
f = np.array(f_pil)
plt.imshow(f, cmap='gray')
plt.title('Imagem original')
```

Let's now try to segment the *cookies*, that is, separate them from the background, so that all pixels belonging to cookies receive one value and pixels that do not belong to cookies (background) receive another value. We will use the *threshold_otsu* and *otsu* functions from the *skiimage* library.

```python
# Searching for an optimal threshold for segmentation, using the Otsu algorithm (global)

t_glob_otsu = threshold_otsu(f) # t_glob_otsu is the threshold (scalar)
glob_otsu = (f >= t_glob_otsu)

# Searching for segmentation, using Otsu algorithm (local) - local thresholds
# In this case it is necessary to choose the size of the neighborhood that will be used in the local search

radius = 10
viz = mm.disk(radius)

t_loc_otsu = otsu(f, viz) # t_loc_otsu is an image (one threshold for each pixel)
loc_otsu = (f >= t_loc_otsu)

# Visualizing the results, as binary image and as segmented (labeled) image

plt.figure(1, figsize=(14,8))

plt.subplot(2,2,1)
plt.imshow(glob_otsu, cmap='gray')
plt.title('Thresholded image - Otsu Global')
plt.subplot(2,2,2)
plt.imshow(glob_otsu)
plt.title('Segmented image (labeled) - Otsu Global')
plt.subplot(2,2,3)
plt.imshow(loc_otsu, cmap='gray')
plt.title('Thresholded image - Otsu Local')
plt.subplot(2,2,4)
plt.imshow(loc_otsu)
plt.title('Segmented image (labeled) - Otsu Local')
```

### Segmentation using the Watershed method

We can solve the same *cookie* segmentation problem by now applying the Watershed method. The first attempt would be to use the watershed method and choose the markers with the help of the histogram. We will use the *watershed* function from the *skiimage.segmentation* library

```python
nbins = 256
hr, bin_edges = np.histogram(f, nbins,(0,255))
plt.plot(hr)
```

The markers for *foreground* and *background* will be chosen as close to the 2 peaks of the histogram. The values do not need to be exact, as they will not be used as a threshold, only as a seed.

```python
markers = np.zeros_like(f)
foreground, background = 1, 2 # labels for the seeds (object = 1 and background = 2)
markers[f < 30.0] = background # all pixels with values <30 will receive the label 2 in the markers image
markers[f > 140.0] = foreground # all pixels with values >140 will receive the label 1 in the markers image

# Calculation of watershed from image and markers
w1 = watershed(f, markers)
plt.figure(1, figsize=(14,8))

plt.subplot(1,3,1)
plt.imshow(f, cmap='gray')
plt.title('Original Image')
plt.subplot(1,3,2)
plt.imshow(w1)
plt.title('Image labeled by Watershed')
plt.subplot(1,3,3)
plt.imshow(mark_boundaries(f,w1))
plt.title('Segmented image - watershed line')
```

The segmentation result obtained was not satisfactory, as we used the original image as watersehd input. Normally the image gradient is used as input (gradient = discontinuities). In this case, we will use the Sobel edge detector to obtain this gradient and then we will run the watershed algorithm on this gradient.

```python
s = sobel(f)
plt.imshow(s,cmap='gray')
plt.title('Edge detection by Sobel')
```

```python
w2 = watershed(s,markers)
plt.figure(1, figsize=(14,8))

plt.subplot(1,3,1)
plt.imshow(f, cmap='gray')
plt.title('Original Image')
plt.subplot(1,3,2)
plt.imshow(w2)
plt.title('Image labeled by Watershed')
plt.subplot(1,3,3)
plt.imshow(mark_boundaries(f,w2))
plt.title('Segmented image - watershed line')
```

The result obtained now with segmentation seems much closer to what was desired. Finally, let's compare the results of the 2 segmentations - by *Otsu* and by *Watershed* - and check how similar they are. We can make this comparison in two ways: qualitative and quantitative. Qualitative comparison is typically done by viewing the segmentations side-by-side or overlapping.

```python
glob_otsu = glob_otsu.astype(int)
w2 = w2.astype(int)

# Ensure both arrays have the same shape
if glob_otsu.shape != w2.shape:
    glob_otsu = glob_otsu[:w2.shape[0], :w2.shape[1]]
```

```python

segj = join_segmentations(glob_otsu, w2)

plt.figure(1, figsize=(14,8))

plt.subplot(1,3,1)
plt.imshow(mark_boundaries(f,glob_otsu))
plt.title('Targeting by Otsu')
plt.subplot(1,3,2)
plt.imshow(mark_boundaries(f,w2))
plt.title('Watershed Segmentation')
plt.subplot(1,3,3)
plt.imshow(segj)
plt.title('Overlap of the 2 segmentations')
```
