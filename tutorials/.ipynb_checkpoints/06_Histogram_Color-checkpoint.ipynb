# Color image transformation

## Pseudocolor transformation

Also known as *density slicing*, it converts a gray level image into a color image, mapping each gray level or range of levels to a different color. This type of mapping is used when the original image has several objects of interest with little variation in gray levels between them.

The following examples show examples of pseudocolor transformation, where the image is mapped into color tables so that each gray level of the original image is mapped into three images R, G, B in the RGB color model.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import sys,os
```

```python
f = mpimg.imread('../figures/retina.tif')
plt.imshow(f)
```

```python
plt.imshow(f,cmap='brg')
```

```python
plt.imshow(f, cmap='hsv')
```

```python
plt.imshow(f,cmap='seismic')
```
